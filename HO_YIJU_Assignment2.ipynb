**Question 10:**
**Top 5 highiet OBP(on bas percentage) with at least 500 PAs in 1977?**

```python
# OBP = (H + BB + HBP) / (AB + BB + HBP + SF)
# PAs = AB + BB + HBP + SH + SF
H = df["H"]
BB = df["BB"]
HBP = df["HBP"]
AB = df["AB"]
SF = df["SF"]
SH = df["SH"]
OBP = (H+BB+HBP)/(AB+BB+HBP+SF)
df["OBP"] = OBP
PA = (AB + BB + HBP + SH + SF)
df["PA"] = PA
df10 = pd.DataFrame(df[(df["PA"]>=500) & (df["yearID"]==1977)][["nameFirst","nameLast","OBP","PA"]].sort_values(by="OBP", ascending = False).reset_index())
del df10["index"]
df10.head()
# Rod Carew, Ken Singleton, Reggie Smith, Mike Hargrove, Joe Morgan.
```

** Graph total HRs per season using bar graph.**

```python
# Bonus1
import matplotlib.pyplot as plt
%matplotlib inline

df1 = df.groupby("yearID")["HR"].sum()
df1

plt.figure(figsize=(20, 8))
pplot = df1.plot(kind = "bar")
pplot.set_title("HR for seasons", fontsize = 30)
pplot.set_xlabel("seasons", fontsize = 20)
pplot.set_ylabel("HR", fontsize = 20)
```
